```python
x = np.linspace(-2, 2, 100)
f = x**2
df = np.diff(f)
dx = np.diff(x)

plt.plot(x, f, label='f')
plt.plot(x[:-1], df/dx, label='df/dx')

plt.axis([-2, 2, -1, 2])
plt.legend()
plt.grid()
plt.show()
```

```python
coefs = [3, 2, -1]

roots = np.roots(coefs)
```

```python
print(roots)
```

```python
x = sym.symbols('x')

pn = 3*x**2 + 2*x - 1

for i in roots:
    display(Math('\\text {At } x=%g, \\quad %s = %g' % (i, sym.latex(pn), pn.subs(x, i))))
```

```python
for i in range(1, 11):
    coefs = np.arange(1, i+1)
    roots = np.roots(coefs)
    print(f'A degree {i-1} polynomial has {len(roots)} roots.')
```

```python
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
```

```python
a = 3
b = 7
c = 5

quadeqP = (-b + sp.sqrt(b**2 - 4*a*c)) / (2*a)
quadeqN = (-b - sp.sqrt(b**2 - 4*a*c)) / (2*a)

print(quadeqP)
print(quadeqN)
```

```python
def quadeq(a, b, c):
    
    out = np.zeros(2)
    
    out[0] = (-b + sp.sqrt(b**2 - 4*a*c)) / (2*a)
    out[1] = (-b - sp.sqrt(b**2 - 4*a*c)) / (2*a)
    
    return out
```

```python
quadeq(2, 7, 5)
```

```python
a = 1
b = range(-5, 6)
c = range(-2, 11)

M = np.zeros((len(b), len(c)))

for bi in range(0, len(b)):
    for ci in range(0, len(c)):
        M[bi, ci] = quadeq(a, b[bi], c[ci])[0]
        
plt.imshow(M, extent=[c[0], c[-1], b[0], b[-1]])
plt.xlabel('c')
plt.ylabel('b')
plt.title('a=' + str(a))
plt.colorbar()
plt.show()
```
