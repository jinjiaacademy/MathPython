```python
x = [-4, 5, 2, 4, 5, 6, 3]
y = [0, 3, 4, 5, 2,-3, 4]

for i in range(len(x)):
    plt.plot(x[i], y[i], 'o', label='point %i' %(i+1))

plt.axis('square')
plt.grid()
plt.legend()
plt.show()
```

```python
import sympy as sym

x, y = sym.symbols('x y')

for x in range(-10, 11):
    y = x**2 - 3*x
    plt.plot(x, y, 'o')
    
plt.xlabel('x')
plt.ylabel('y=x**2 - 3*x')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
p1 = [-3, -1]
p2 = [4, 4]

plt.plot([p1[0],p2[0]], [p1[1], p2[1]], color=[1, .7, .2], linewidth=5)

plt.axis('square')
plt.axis([-6, 6, -6, 6])

plt.show()
```

```python

x = 3
y = 5

plt.plot(x, y, 'ro')
plt.plot([0, x], [0, y], 'r')

plt.axis('square')
plt.axis([-6, 6, -6, 6])
plt.grid()

plt.plot([-6, 6], [0, 0], 'k')
plt.plot([0, 0], [-6, 6], 'k')

plt.show()
```

```python
import math

for x in range(-20, 21):
    y = math.sqrt(abs(x))
    plt.plot([0,x], [0,y])

plt.show()
```

```python
plt.plot([0, 2], [0, 0], 'k')
plt.plot([0, 0], [0, 2], 'g')
plt.plot([0, 2], [2, 2], 'r')
plt.plot([2, 2], [0, 2], 'm')

plt.axis('square')
plt.axis([-3, 5, -3, 5])
plt.title('Colorful Square')

plt.show()
```

```python
# y = mx + b

x = [-5, 5]
m = -2
b = .1

# y = [340, 2345]
# for i in range(len(x)):
#     y[i] = m*x[i] + b

y = m*np.array(x) + b

plt.plot(x, y, label='y=%sx+%s' % (m, b))

plt.axis('square')
plt.grid()
plt.xlim(x)
plt.ylim(x)

axis = plt.gca()
plt.plot(axis.get_xlim(), [0, 0], 'k--')
plt.plot([0, 0], axis.get_ylim(), 'k--')

plt.legend()

plt.show()
```

```python
x = [-5, 5]

m = [.7, -1.25]
b = [-2, .75]

for i in range(len(m)):
    y = m[i]*np.array(x)+b[i]
    plt.plot(x, y, label='y=%sx + %s' % (m[i], b[i]))

plt.axis('square')
plt.grid()
plt.xlim(x)
plt.ylim(x)

axis = plt.gca()
plt.plot(axis.get_xlim(), [0, 0], 'k--')
plt.plot([0, 0], axis.get_ylim(), 'k--')

plt.xlabel('x')
plt.ylabel('y')
plt.title('The plot')
plt.legend()

plt.show()
```

```python
x = range(-3, 4)
y = np.zeros(len(x))

for i in range(len(x)):
    y[i] = 2 - x[i]**2

plt.plot(x, y, 's-')
plt.show()
```

```python
x = np.linspace(-3, 4, 10)
y = 2 + np.sqrt(abs(x))

plt.plot(x, y, 'm-')
plt.show()
```

```python
x = np.linspace(-4, 4, 40)
pwrs = range(-1, 4)

for pwr in pwrs:
    y = x**pwr
    plt.plot(x, y, '-',label='$y = x**{%s}$'%pwr, linewidth=5)
    
plt.ylim([-20, 20])

plt.xlabel('x')
plt.ylabel('y')
plt.legend()
    
plt.show()
```
