```python
import matplotlib.pyplot as plt
```

```python
x = 3
y = 5

plt.plot(x, y, 'kp')

plt.axis('square')
plt.axis([-6, 6, -6, 6])
plt.grid()

plt.show()
```

```python
x = [-4, 5, 2, 4, 5, 6, 3]
y = [0, 3, 4, 5, 2,-3, 4]

for i in range(len(x)):
    plt.plot(x[i], y[i], 'o', label='point %i' %(i+1))

plt.axis('square')
plt.grid()
plt.legend()
plt.show()
```

```python
plt.plot(4, 2, 'r+')

axis = plt.gca()
ylim = axis.get_ylim()
axis.set_ylim([ylim[0], 5.39485739])
plt.show()
```

```python
import sympy as sym

x, y = sym.symbols('x y')

for x in range(-10, 11):
    y = x**2 - 3*x
    plt.plot(x, y, 'o')
    
plt.xlabel('x')
plt.ylabel('y=x**2 - 3*x')
plt.show()
```
