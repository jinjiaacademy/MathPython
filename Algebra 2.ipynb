```python
def quadeq(a, b, c):
    
    out = sp.zeros(2, dtype=complex)
    
    out[0] = (-b + sp.sqrt(b**2 - 4*a*c)) / (2*a)
    out[1] = (-b - sp.sqrt(b**2 - 4*a*c)) / (2*a)
    
    return out
```

```python
a = 1
b = range(-5, 6)
c = range(-2, 11)

M = np.zeros((len(b), len(c)), dtype=complex)

for i in range(0, len(b)):
    for j in range(0, len(c)):
        M[i, j] = quadeq(a, b[i], c[j])[0]
        
plt.subplot(1, 3, 1)
plt.imshow(np.real(M), extent=[c[0], c[-1], b[0], b[-1]])
plt.axis('on')
plt.title('Real Part')

plt.subplot(1, 3, 2)
plt.imshow(np.imag(M), extent=[c[0], c[-1], b[0], b[-1]])
plt.axis('off')
plt.title('Imag Part')

plt.subplot(1, 3, 3)
plt.imshow(np.absolute(M), extent=[c[0], c[-1], b[0], b[-1]])
plt.axis('off')
plt.title('Magnitude')


plt.show()
```

```python
x = np.linspace(0, 2*np.pi, 20)
fx = -(np.cos(x) + np.sqrt(x))

fmax = np.max(fx)
fmaxidx = np.argmax(fx)
print(fmax, fmaxidx)

plt.plot(x, fx, 'bo-')
plt.plot(x[fmaxidx], fmax, 'rs')

plt.show()
```

```python
x = np.linspace(-2, 2, 51)

fx = x**3 + x**4

fminidx = np.argmin(fx)


plt.plot(x, fx, 'bs-')
plt.plot(x[fminidx], fx[fminidx], 'rs')
plt.xlim([-1.2, .5])
plt.ylim([-.12, .2])

plt.show()
```

```python
x = np.linspace(0, 12*np.pi, 213)
fx = -np.cos(x) - np.sqrt(x)

fpeaks = ss.find_peaks(fx)
print(fpeaks[0])

plt.plot(x, fx, 'b-')
plt.plot(x[fpeaks[0]], fx[fpeaks[0]], 'o')
plt.show()
```
