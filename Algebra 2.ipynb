```python
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
```

```python
a = 3
b = 7
c = 5

quadeqP = (-b + sp.sqrt(b**2 - 4*a*c)) / (2*a)
quadeqN = (-b - sp.sqrt(b**2 - 4*a*c)) / (2*a)

print(quadeqP)
print(quadeqN)
```

```python
def quadeq(a, b, c):
    
    out = np.zeros(2)
    
    out[0] = (-b + sp.sqrt(b**2 - 4*a*c)) / (2*a)
    out[1] = (-b - sp.sqrt(b**2 - 4*a*c)) / (2*a)
    
    return out
```

```python
quadeq(2, 7, 5)
```

```python
a = 1
b = range(-5, 6)
c = range(-2, 11)

M = np.zeros((len(b), len(c)))

for bi in range(0, len(b)):
    for ci in range(0, len(c)):
        M[bi, ci] = quadeq(a, b[bi], c[ci])[0]
        
plt.imshow(M, extent=[c[0], c[-1], b[0], b[-1]])
plt.xlabel('c')
plt.ylabel('b')
plt.title('a=' + str(a))
plt.colorbar()
plt.show()
```

```python
def quadeq(a, b, c):
    
    out = sp.zeros(2, dtype=complex)
    
    out[0] = (-b + sp.sqrt(b**2 - 4*a*c)) / (2*a)
    out[1] = (-b - sp.sqrt(b**2 - 4*a*c)) / (2*a)
    
    return out
```

```python
a = 1
b = range(-5, 6)
c = range(-2, 11)

M = np.zeros((len(b), len(c)), dtype=complex)

for i in range(0, len(b)):
    for j in range(0, len(c)):
        M[i, j] = quadeq(a, b[i], c[j])[0]
        
plt.subplot(1, 3, 1)
plt.imshow(np.real(M), extent=[c[0], c[-1], b[0], b[-1]])
plt.axis('on')
plt.title('Real Part')

plt.subplot(1, 3, 2)
plt.imshow(np.imag(M), extent=[c[0], c[-1], b[0], b[-1]])
plt.axis('off')
plt.title('Imag Part')

plt.subplot(1, 3, 3)
plt.imshow(np.absolute(M), extent=[c[0], c[-1], b[0], b[-1]])
plt.axis('off')
plt.title('Magnitude')


plt.show()
```
