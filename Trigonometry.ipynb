```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
nums = np.random.rand(1000)

minval = 2
maxval = 17

nums = nums*(maxval-minval) + minval

plt.plot(nums, 's')
plt.show()

plt.hist(nums)
plt.show()
```

```python
nums = np.random.randn(1000)

plt.plot(nums, 's')
plt.show()

plt.hist(nums)
plt.show()
```
